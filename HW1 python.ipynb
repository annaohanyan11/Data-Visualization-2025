```python
import pandas as pd
```

```python
data = pd.read_csv("crime_data.csv")
data.head()
```

```python
data.isna().sum()
```

```python
missing50 = data.isna().sum()[data.isna().sum() > (0.5 * len(data))].index
data_cleaned = data.drop(columns=missing50)
print("Columns dropped:", missing50)
```

```python
data_cleaned['DATE OCC'] = pd.to_datetime(data_cleaned['DATE OCC'])
data_cleaned['Year'] = data_cleaned['DATE OCC'].dt.year
data_cleaned['Month'] = data_cleaned['DATE OCC'].dt.month
data_cleaned['Day'] = data_cleaned['DATE OCC'].dt.day
```

```python
data_cleaned['Hour'] = data_cleaned['TIME OCC'].astype(str).str.zfill(4).str[:2].astype(int)
data_cleaned.head()
```

```python
data_filtered = data_cleaned[(data_cleaned['Year'] == 2023) & (data_cleaned['Crm Cd Desc'] == 'BURGLARY')]
data_cleaned[(data_cleaned['Year'] == 2023)]
```

```python
grouped = data_filtered.groupby('AREA NAME').agg(
    Total_Crimes=('Crm Cd Desc', 'count'),
    Avg_Victim_Age=('Vict Age', 'mean')
).sort_values(by='Total_Crimes', ascending=False)
grouped
```
